```python
from matplotlib import pyplot as plt

x = [1,1.5,2,2.5,3]
y = [10,8,9,10.5,12]

plt.plot(x,y)

plt.title("Line graph")
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.show()
```
