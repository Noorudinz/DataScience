```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

Plotting a vector

```python
plt.quiver(0,0,2,6)
plt.show()
```

```python
plt.quiver(0,0,2,10, scale_units='xy', angles='xy', scale=1)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

```python
plt.quiver(0,0,-5,10, scale_units='xy', angles='xy', scale=1, color='blue')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Addition of two vectors

```python
vector_1 = np.asarray([0,0,2,6])

vector_2 = np.asarray([0,0,-3,2])

sum = vector_1 + vector_2
print(sum)
```

```python
plt.quiver(0,0,2,6, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-3,2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,-1,8, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

Subtraction of two vector

```python
vector_1 = np.asarray([0,0,5,3])

vector_2 = np.asarray([0,0,-4,-2])

difference = vector_1 - vector_2
print(difference)
```

```python
plt.quiver(0,0,5,3, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-4,-2, scale_units='xy', angles='xy', scale=1, color='y')
plt.quiver(0,0,9,5, scale_units='xy', angles='xy', scale=1, color='r')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```

Multiplying a vector by a scalar

```python
vector_1 = np.asarray([0,0,5,10])

vector_2 = 2*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,5,10, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,10,20, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-20,20)
plt.ylim(-20,20)
plt.show()
```

```python
vector_1 = np.asarray([0,0,5,10])

vector_2 = -0.5*vector_1

print(vector_2)
```

```python
plt.quiver(0,0,5,10, scale_units='xy', angles='xy', scale=1, color='b')
plt.quiver(0,0,-2.5,-5, scale_units='xy', angles='xy', scale=1, color='y')
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.show()
```
